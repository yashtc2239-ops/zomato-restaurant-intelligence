```python
import pandas as pd
import numpy as np

# Load both datasets
df_bengaluru = pd.read_csv('../data/raw/Zomato_bengaluru.csv')
df_global = pd.read_csv('../data/raw/Zomato_global.csv', encoding='latin-1')

print("✅ Both datasets loaded successfully")
```

```python
print("=" * 50)
print("DATASET 1: BENGALURU")
print("=" * 50)
print(f"Shape        : {df_bengaluru.shape}")
print(f"\nColumns      :\n{df_bengaluru.columns.tolist()}")
print(f"\nData Types   :\n{df_bengaluru.dtypes}")
print(f"\nNull Counts  :\n{df_bengaluru.isnull().sum()}")
print(f"\nFirst 3 Rows :\n{df_bengaluru.head(3)}")
```

```python
print("=" * 50)
print("DATASET 2: GLOBAL")
print("=" * 50)
print(f"Shape        : {df_global.shape}")
print(f"\nColumns      :\n{df_global.columns.tolist()}")
print(f"\nData Types   :\n{df_global.dtypes}")
print(f"\nNull Counts  :\n{df_global.isnull().sum()}")
print(f"\nFirst 3 Rows :\n{df_global.head(3)}")
```

```python
# CHECK 1 — All unique rate values
print("All unique rate values:")
print(df_bengaluru['rate'].unique())
print(f"\nTotal unique rate values: {df_bengaluru['rate'].nunique()}")

# CHECK 2 — Null percentage
print("\nBengaluru Null %:")
null_pct = (df_bengaluru.isnull().sum() / len(df_bengaluru) * 100).round(2)
print(null_pct[null_pct > 0])

# CHECK 3 — Country distribution
print("\nCountry Code distribution:")
print(df_global['Country Code'].value_counts())

# CHECK 4 — Duplicates
print(f"\nBengaluru duplicates: {df_bengaluru.duplicated().sum()}")
print(f"Global duplicates   : {df_global.duplicated().sum()}")
```

```python
# CHECK 5 — Cost column dirty values
print("Sample cost values:")
print(df_bengaluru['approx_cost(for two people)'].unique()[:20])

# CHECK 6 — Online order & book table distribution
print("\nOnline Order distribution:")
print(df_bengaluru['online_order'].value_counts())

print("\nBook Table distribution:")
print(df_bengaluru['book_table'].value_counts())

# CHECK 7 — Location top 10
print("\nTop 10 Locations:")
print(df_bengaluru['location'].value_counts().head(10))
```
